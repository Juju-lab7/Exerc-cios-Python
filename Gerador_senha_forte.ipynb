```
#perguntar o tamanho da senha
#criar lista de caracteres possíveis
#gerar caracteres aleatorios
#montar a senha
#mostrar a senha

import random
import string

def gerar_senha(tamanho):
  caracteres = string.ascii_letters + string.digits + string.punctuation
  senha = ''.join(random.choice(caracteres) for i in range(tamanho))
  return senha
  # A linha abaixo nunca será alcançada devido ao 'return' anterior.
  # print(gerar_senha(15))

def gerar_senha_forte(tamanho, maiusculas=True, minusculas=True, numeros=True, simbolos=True):
  caracteres = ''
  if maiusculas:
    caracteres += string.ascii_uppercase
  if minusculas:
    caracteres += string.ascii_lowercase
  if numeros:
    caracteres += string.digits
  if simbolos:
    caracteres += string.punctuation

  # Certifica-se de que há pelo menos um tipo de caractere para gerar a senha
  if not caracteres:
    print("Pelo menos um tipo de caractere deve ser selecionado.")
    return ""

  senha = ''.join(random.choice(caracteres) for i in range(tamanho))
  return senha

def main():
  tamanho = int(input('Qual o tamanho da senha?: '))

  incluir_maiusculas = input('Incluir letras maiúsculas? (s/n): ').upper() == 'S'
  incluir_minusculas = input('Incluir letras minúsculas? (s/n): ').upper() == 'S'
  incluir_numeros = input('Incluir numeros? (s/n): ').upper() == 'S'
  incluir_simbolos = input('Incluir símbolos? (s/n): ').upper() == 'S'

  # A função 'gerar_senha' original aceita apenas um argumento.
  # Para usar as opções de força, devemos chamar 'gerar_senha_forte'.
  senha = gerar_senha_forte(tamanho, incluir_maiusculas, incluir_minusculas, incluir_numeros, incluir_simbolos)
  print(f'A senha gerada: {senha}')

if __name__ == '__main__':
  main()
```
